```python
import numpy as np
import matplotlib.pyplot as plt  
from mpl_toolkits import mplot3d
```

```python
E0 = 1
B0 = 1
q = 1
m = 5

wx = q*B0/m

def x(t): return -(E0/B0)*(1/wx)*(np.cos(wx*t)-1)
def y(t): return (E0/B0)*((1/wx)*np.sin(wx*t)-t)

ts = np.arange(0, 20.1*np.pi, 0.1)
xs = [x(t) for t in ts]
ys = [y(t) for t in ts]

fig, ax = plt.subplots()
handles, labels = ax.get_legend_handles_labels()
ax.legend(handles, labels)

ax.plot(xs,ys)
ax.set_title(r'Trayectoría de la carga ($q={}$C, $m={}$kg, $t={}$seg, $E0={}$, $B={}$)'.format(q,m,ts[-1],B0,E0))
ax.set_ylabel('y(t)')
ax.set_xlabel('x(t)')
plt.show()
```

```python
zs = [10*t for t in ts]
ax = plt.axes(projection='3d')
ax.plot3D(xs, ys, zs)
ax.set_xlabel('x(t)')
ax.set_ylabel('y(t)')
ax.set_zlabel('z(t)')
ax.set_title(r'Trayectoría de la carga $z(t)=v_0 t$')
```
